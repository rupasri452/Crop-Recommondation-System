```python
# Separate features and target
y = data['label']
x = data.drop(['label'], axis=1)

print("Shape of x", x.shape)
print("Shape of y", y.shape)

# Split the dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0
)

print("The shape of x train", x_train.shape)
print("The shape of x test", x_test.shape)
print("The shape of y train", y_train.shape)
print("The shape of y test", y_test.shape)
```

```python
# Elbow Method

wcss = []

for i in range(1, 11):
    kmeans = KMeans(
        n_clusters=i,
        init='k-means++',
        random_state=0
    )

    kmeans.fit(x)

    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

```python
# Apply K-Means Clustering
km = KMeans(
    n_clusters=4,
    init='k-means++',
    max_iter=300,
    n_init=10,
    random_state=0
)

y_means = km.fit_predict(x)

# Create a DataFrame with cluster labels
a = data['label']
y_means = pd.DataFrame(y_means)

z = pd.concat([y_means, a], axis=1)
z = z.rename(columns={0: 'cluster'})

print("Let's check the results after applying the K-Means clustering analysis\n")

print("Crops in First Cluster:")
print(z[z['cluster'] == 0]['label'].unique().tolist())

print("\n--------------------------------------")

print("Crops in Second Cluster:")
print(z[z['cluster'] == 1]['label'].unique().tolist())

print("\n--------------------------------------")

print("Crops in Third Cluster:")
print(z[z['cluster'] == 2]['label'].unique().tolist())

print("\n--------------------------------------")

print("Crops in Fourth Cluster:")
print(z[z['cluster'] == 3]['label'].unique().tolist())
```

```python
from sklearn.linear_model import LogisticRegression

# Create the model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(x_train, y_train)

# Predict the test data
y_pred = model.predict(x_test)

# Display the model
print(model)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import pickle

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy :", accuracy)

# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:\n")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Save Model
pickle.dump(model, open("model.pkl", "wb"))

print("\nModel Saved Successfully")
```
